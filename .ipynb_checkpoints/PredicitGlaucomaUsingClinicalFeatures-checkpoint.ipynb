### import Libraries

```python
from ExtractClinicalFeatures import detectroi
from ExtractClinicalFeatures import plot_image
from ExtractClinicalFeatures import segmentdiscandcup
from ExtractClinicalFeatures import calculateCDR
from ExtractClinicalFeatures import calculateNRR

import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
import random
import imgaug as ia
```

```python
from sklearn import metrics
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
```

### Initialize Variables

```python
basepath = "data"
roipath = "FullROI"
orgpath = "OrgImage"
labelpath = "label"
imgreadpath = os.path.join(basepath , orgpath)
lblreadpath = os.path.join(basepath , labelpath)
roireadpath = os.path.join(basepath , roipath)
images = os.listdir(imgreadpath)
```

### Detect ROI from Fundus Image

```python
##Scan all images and detect the Region on Interest and save in ROI folder
for fname in images:
        filepath = os.path.join(imgreadpath, fname)
        #print(filepath)
        writepath = os.path.join(basepath,roipath,fname)   
        # skip directories
        if os.path.isdir(filepath):
              continue
        img = cv2.imread(filepath,1)
        img_roi = detectroi(img)
        cv2.imwrite(writepath,img_roi)
        #print(writepath)
```

### Segment Disc and Cup and Calculate CDR (Cup to Disc Ratio) and NRR (Neuroretinal Rim)

```python
#Segment ROI image into cup, disc and cupdisc 
#Calcuate CDR
#Calcuate NRR
data = []
filepath = os.path.join(basepath, roipath)
images = os.listdir(filepath)
for fname in images:
    imagepath = filepath+"\\"+fname
    print("Processing Image - " + fname)

    lblfilereadpath = os.path.join(lblreadpath, os.path.splitext(fname)[0]+'.txt')
    f = open(lblfilereadpath, "r")
    label = f.read()
    f.close()
    image = cv2.imread(imagepath,1)
    segmentdiscandcup(image)
    cup = cv2.imread("cup.png",0)
    disc = cv2.imread("disc.png",0)
    cdr_cal = calculateCDR(cup,disc,False)
    nrr_cal = calculateNRR('cupdisc.png',False)
  
    os.remove('cup.png')
    os.remove('disc.png')
    os.remove('cupdisc.png')
    os.remove('inferior.png')
    os.remove('nasal.png')
    os.remove('temporal.png')   
    os.remove('superior.png')
    data.append([fname, cdr_cal, nrr_cal,int(label)])


df = pd.DataFrame(data, columns=['Image','cdr', 'nrr','label'])
df.to_csv('clinicalfeaturesrevised.csv') 
```

## Modelling Approach

## Logistic Regression

```python
####Load the Dataset 
cfdf = pd.read_csv("clinicalfeaturesrevised.csv")
cfdf.head()
```

```python
cfdf=cfdf.drop(['Image'],axis=1)
```

```python
glnglratio = (sum(cfdf['label'])/len(cfdf['label']))*100
glnglratio
```

```python
ngl_df = pd.DataFrame(cfdf.loc[(cfdf.label ==  0)])
len(ngl_df)
```

```python
gl_df = pd.DataFrame(cfdf.loc[(cfdf.label ==  1)])
len(gl_df)
```

```python
X=cfdf.drop(['label'],axis=1)
y=cfdf['label']
y.describe()
```

```python
#### Perform Normalization 
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size = 0.3, random_state = 4)
sc=StandardScaler()
Xtrain=pd.DataFrame(sc.fit_transform(X_train),columns=X.columns)
Xtest=pd.DataFrame(sc.transform(X_test),columns=X.columns)
```

```python
# LR without hyperparameter tunning
log_updated= LogisticRegression(class_weight='balanced')
 
log_updated.fit(Xtrain, y_train)
predictions = log_updated.predict(Xtest)
probas=log_updated.predict_proba(Xtest)
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
print("Accuracy->" ,accuracy_score(y_test,predictions))
print("AUC->",metrics.roc_auc_score(y_test,probas[:, 1]))
```

## Random Forest

```python
rfc = RandomForestClassifier(class_weight='balanced',random_state=4)
rfc.fit(Xtrain,y_train)
predictions = rfc.predict(Xtest)
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
print("Accuracy->", accuracy_score(y_test,predictions))
print("AUC->",metrics.roc_auc_score(y_test,probas[:, 1]))
```

## XgBoosting

```python
import xgboost as xgb
parameters = {'max_depth': range(3, 12, 3)}

xgb = xgb.XGBClassifier(random_state=4)

gbm_f1_cv1 = GridSearchCV(xgb, 
                   parameters, 
                   cv=StratifiedKFold(n_splits=4), 
                   scoring="accuracy",
                   verbose = 1)
gbm_f1_cv1.fit(X_train, y_train)


print(gbm_f1_cv1.best_params_)
```

```python
import xgboost as xgb
parameters = {'scale_pos_weight': [1,3,5,7]}

xgb = xgb.XGBClassifier(random_state=4)

gbm_f1_cv2 = GridSearchCV(xgb, 
                   parameters, 
                   cv=StratifiedKFold(n_splits=4), 
                   scoring="accuracy",
                   verbose = 1)
gbm_f1_cv2.fit(X_train, y_train)

print(gbm_f1_cv2.best_params_)
```

```python
import xgboost as xgb
parameters = {'subsample': [0.6,0.8,1]}

xgb = xgb.XGBClassifier(random_state=4)

gbm_f1_cv3 = GridSearchCV(xgb, 
                   parameters, 
                   cv=StratifiedKFold(n_splits=5), 
                   scoring="accuracy",
                   verbose = 1)
gbm_f1_cv3.fit(X_train, y_train)
print(gbm_f1_cv3.best_params_)
```

```python
import xgboost as xgb
gbm_f1 = xgb.XGBClassifier(subsample=0.6 ,scale_pos_weight=1, max_depth=3,random_state=4)

gbm_f1.fit(X_train,y_train)

predictions = gbm_f1.predict(X_test)
probas=gbm_f1.predict_proba(X_test)
print(classification_report(y_test,predictions))
print(confusion_matrix(y_test,predictions))
print("Accuracy->", accuracy_score(y_test,predictions))
print("AUC->",metrics.roc_auc_score(y_test,probas[:, 1]))
```
